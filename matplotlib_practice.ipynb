```python
# import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot using object oriented interface method. 

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plot()

ax = plt.axes()
ax.plot(x_axis, y_axis)

# To change figure-level attributes (such as axix labels, a title, or a legend) or save the figure as an image, use 
# fig= plt.figure()
```

```python
# SCATTER PLOTS
# Using plt.plot() method

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# using plt.scatter()

#plt.scatter(x_axis, y_axis, c='red', label='Chicago')
plt.scatter(y_axis, x_axis, c='red', label='Chicago', marker='x', s=y_axis)
plt.gca().invert_yaxis()
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
# CREATING BUBBLE CHARTS 

plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# making the bubbles a liilte more large so it is visible
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# PYTHONS LIST COMPREHENSIONS **** IMPORTANT
#new_list = [expression for item in list if conditional]

plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# USING OBJECT ORIENTED INTERFACE

fig, ax= plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# SKILL DRILL

fig, ax= plt.subplots()
ax.scatter(y_axis, x_axis, c='skyblue', label ='Boston', alpha=0.8, edgecolors='black', linewidth=2, s=[i*5 for i in y_axis])
ax.invert_yaxis()
ax.set_title('PyBer Fare by Month')
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
ax.set_xlim(0,50)
ax.legend()
```

```python
# CREATING A PIE CHART

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# ASSIGNING COLORS
colors = ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen', 'greenyellow', 'orange', 'gold', 'indianred', 'tomato', 'mistyrose' ]

# EXPLODE OR HIGHLIGHT THE INTERESTED PIE SLICE
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# INCREASING PIE CHART SIZE TO A READABLE AND SPACED OUT
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')

plt.show()
```

```python
# USING OBJECT ORIENTED INTERFACE

fig, ax = plt.subplots(figsize=(8,8))
colors_1 = ['powderblue', 'violet', 'dodgerblue', 'green', 'palegoldenrod', 'greenyellow', 'orange', 'gold', 'indianred', 'tomato', 'mistyrose','mediumspringgreen' ]
explode_values_1 = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
       labels = x_axis,
       explode= explode_values_1,
       colors=colors_1,
       labeldistance = 1.2,
       shadow=True,
       startangle=90,
       autopct = '%.1f%%',
       counterclock=True,)
ax.invert_xaxis()
plt.show()
```

```python
# ADDING ERROR BARS
%matplotlib inline

#import dependencies.
import matplotlib.pyplot as plt
import statistics

# plt.errorbar(x_axis, y_axis, yerr=<)
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y_axix
stdev = statistics.stdev(y_axis)
stdev
```

```python
# error bar
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
# making error bars scalar and adding acapsize of 3 to the error bars

plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Object oriented interface
fig, ax=plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# adding error bars to bar graphs

plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax=plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure.
fig, ax=plt.subplots(figsize=(8,8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0,51, step=5.0))

# create minor ticks at imcrement of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
